```python
#import
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import numpy as np
from collections import defaultdict

#Definition
ConcentrationA=[]
ConcentrationB=[]
ConcentrationC=[]
ConcentrationD=[]
ConcentrationE=[]

#Here to change the interval length of Time range
Step= 0.0025

#Here to change the initial concentration of the reactants
cA0=2.239
cB0=2.629
cC0=0
cD0=0
cE0=0

#Here to change the rate constant K1,K-1, K2
K1= 0.249
Krev= 0.0000238
K2= 0.0134
        
#Here to change the Time Range and its interval(unit is 0.01s)
time = np.arange(0,100,Step)


# Rate1=K1[A][B]   Rate2=K2[C]
# From the Steady State: K-1[C]+ K2[C]= K1[A][B],so [C]= [k1/(K-1+K2)][A][B]
        
for x in time:
    cC0 = Step*K1*cA0*cB0+cC0
    cA0 = cA0-Step*K1*cA0*cB0
    cB0 = cB0-Step*K1*cA0*cB0
    cA0 = cA0+Step*Krev*cC0
    cB0 = cB0+Step*Krev*cC0
    cC0 = cC0-Step*Krev*cC0
    cD0 = Step*K2*cC0+cD0
    cE0 = Step*K2*cC0+cE0
    cC0 = cC0-Step*K2*cC0
    
    ConcentrationD.append(cD0)
    ConcentrationB.append(cB0)
    ConcentrationA.append(cA0)
    ConcentrationC.append(cC0)
    ConcentrationE.append(cE0)

plt.plot(time,ConcentrationA)
plt.plot(time,ConcentrationB)
plt.plot(time,ConcentrationC)
plt.plot(time,ConcentrationD)
plt.plot(time,ConcentrationE)
_ = plt.legend(['A', 'B', 'C','D','E'])
plt.xlabel('Time/s')
plt.ylabel('Concentration/M')
'''plt.savefig('1.png')'''
plt.show()
```
